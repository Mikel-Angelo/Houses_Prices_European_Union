```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'...')
df
```

```python
df.drop(['MEASURE', 'FREQUENCY', 'Flag Codes'], axis = 1, inplace=True)
df.set_index('TIME', inplace=True)
df
```

```python
Value = df.Value
Value
Value = np.resize(Value,(25,21))
Value = Value.transpose() #Change the length of row-column - values ( 21-25=transpose(25-21) = new_df(21-25) )
```

```python
new_df = pd.DataFrame(data=Value, index=df.index.unique(), columns= df['LOCATION'].unique())
new_df
```

```python
plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(18, 12))

ax.plot(new_df['AUT'])
ax.plot(new_df['BEL'])
ax.plot(new_df['CZE'])
ax.plot(new_df['DNK'])
ax.plot(new_df['FIN'])
ax.plot(new_df['FRA'])
ax.plot(new_df['DEU'])
ax.plot(new_df['GRC'])
ax.plot(new_df['HUN'])
ax.plot(new_df['IRL'])
ax.plot(new_df['ITA'])
ax.plot(new_df['LUX'])
ax.plot(new_df['NLD'])
ax.plot(new_df['POL'])
ax.plot(new_df['PRT'])
ax.plot(new_df['SVK'])
ax.plot(new_df['ESP'])
ax.plot(new_df['SWE'])
ax.plot(new_df['EST'])
ax.plot(new_df['LVA'])
ax.plot(new_df['LTU'])
ax.plot(new_df['SVN'])
ax.plot(new_df['OECD'])
ax.plot(new_df['BGR'])
ax.plot(new_df['ROU'])


ax.set_title("Price Houses in European Union")
ax.set_xlabel('YEARS')
ax.set_ylabel('PRICE')

plt.show()
```

```python
for i in new_df:
    fig, ax = plt.subplots(figsize=(18, 12))
    
    ax.plot(new_df[i])
    ax.set_title(i)
    ax.set_xlabel('YEARS')
    ax.set_ylabel('PRICE')
    plt.show()
    
    
```

```python
aver_price = new_df.mean()
plt.figure(figsize=(12,8))
aver_price.plot(kind = "barh", y = "mean", legend = False, alpha=0.3, title = "Average Prices")
```

```python
def fig_plots(name):
    fig, ax = plt.subplots(figsize=(18, 12))
    plt.plot(new_df[name])
    plt.title(name)
    plt.xlabel('Year')
    plt.ylabel('Price')
    
```

```python
fig_plots('GRC')
```
